```sql
Select p.FullName, p.LogonName, p.EmailAddress
-- ,CHARINDEX('@',p.EmailAddress) as POSITION
,LEFT(p.EmailAddress,CHARINDEX('@',p.EmailAddress)-1) as Mailbox
,SUBSTRING(p.EmailAddress, CHARINDEX('@',p.EmailAddress)+1,200) as Domain
From Application.People p
where EmailAddress is not NULL;
```

```sql
Select LEFT(FullName, 1) + '. ' + SUBSTRING(FullName, CHARINDEX(' ', FullName) +1, 200) as Name
From Application.People FullName
--present a listz of people with capitalized initaial first name followed by last
```

```sql
Select LEFT(FullName, 1) + '. ' + SUBSTRING(FullName, CHARINDEX(' ', FullName) +1, 200) as Name,
case
when (IsSalesPerson=1) then 'Sales Person'
when (IsEmployee=1) then 'Employee'
else 'Customer'
end
as [Contact Type]
from Application.People;

-- using people table give count of sales person vs employees vs non-employee
-- sample output: 
-- Sales people EMployee Others
--xx
```

```sql
select count(IIF(p.IsSalesPerson=1, 1, null)) as Sales
,count(IIF(p.IsEmployee=1,1,null)) as Employees
,count(IIF(p.IsEmployee=1 AND p.IsSalesPerson=0,1,null)) as NonSalesEmployees
,count(IIF(p.IsEmployee=0,1,null)) as NonEmployees
from Application.People p;
```
